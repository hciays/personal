# NetworkX

## Setup

````python
# import the NetworkX library
import networkx as nx
````

## Graph
**Graph** consists of **nodes** and **edges**. After creating the graph instance, we will be adding nodes, edges and their attributes to the graph. \\

NetworkX provides several classes for creating different types of graphs (directed and undirected graph) and also multigraphs.

[Different Graph types](https://networkx.org/documentation/stable/reference/classes/index.html) supported by NetworkX. \\

````python
# Instantiating graph
# Create an undirected graph G
G = nx.Graph()
print(G.is_directed())

# Create a directed graph H
H = nx.DiGraph()
print(H.is_directed())

# Add graph level attribute
G.graph["Name"] = "Bar"
G.graph["Att"] = "sddssd"
print(G.graph)
````

## Node

Adding nodes (with attributes) to already created NetworkX graphs.

````python
# Add one node with node level attributes
# nodeid, feature and label
G.add_node(0, feature=0, label=0)


# Get attributes of the node 0
node_0_attr = G.nodes[0]
print("Node 0 has the attributes {}".format(node_0_attr))
````

````python
# Add multiple nodes with attributes
# tuple of attributes
G.add_nodes_from([
  (1, {"feature": 1, "label": 1}),
  (2, {"feature": 2, "label": 2})
])

# Loop through all the nodes
# Set data=True will return node attributes
for node in G.nodes(data=True):
  print(node)

# Get number of nodes
num_nodes = G.number_of_nodes()
print("G has {} nodes".format(num_nodes))
````

## Edge

Edges can also have attributes

````python
# Add one edge with edge weight 0.5
# (source, destination)
G.add_edge(0, 1, weight=0.5)

# Get attributes of the edge (0, 1)
edge_0_1_attr = G.edges[(0, 1)]
print("Edge (0, 1) has the attributes {}".format(edge_0_1_attr))
````

````python
# Add multiple edges with edge weights
G.add_edges_from([
  (1, 2, {"weight": 0.3}),
  (2, 0, {"weight": 0.1})
])

# Loop through all the edges
# Here there is no data=True, so only the edge will be returned
for edge in G.edges():
  print(edge)

# Get number of edges
num_edges = G.number_of_edges()
print("G has {} edges".format(num_edges))
````

## Visualization

````python
# Draw the graph
nx.draw(G, with_labels = True)
````

## Node Properties
**Degree:** The degree of a node is the number of edges connected to the node \\

**Neighbor:** Node that is connected to the target node by an edge


````python
node_id = 0

# Degree of node 1
print("Node {} has degree {}".format(node_id, G.degree[node_id]))

total_neigh = 0
# Get neighbor of node 1
for neighbor in G.neighbors(node_id):
  total_neigh+=1
  print("Node {} has neighbor {}".format(node_id, neighbor))

print(total_neigh)
````

## Centrality and other functionalities

NetworkX also provides several useful methods to study graphs e.g centrality.

````python
num_nodes = 4
# Create a new path like graph and change it to a directed graph
G = nx.DiGraph(nx.path_graph(num_nodes))
nx.draw(G, with_labels = True)

# Get the eigenvector centrality
centrality = nx.eigenvector_centrality(G)
print("centrality", centrality)
````

## Data

Each graph in PyG is represented by a single [`Data`](https://pytorch-geometric.readthedocs.io/en/latest/modules/data.html#torch_geometric.data.Data) object.

The data object holds all the information about the representation of the graph. \\
For example:
```
Data(edge_index=[2, 1144], x=[20, 128], y=[20], train_mask=[20])
```
This `data` object holds 4 attributes:
1.  The `edge_index`: holds the information about the **graph connectivity**, *i.e.*, (source_node, destination_node).

2. The **node features** represented as `x` (each of the 20 nodes is assigned a 128-dim feature vector).

3. The **node labels** represented as `y` (each node is assigned to exactly one class).
4. The **additional attributes** e.g `train_mask`, which describes for which nodes are used for training.


The `data` object also provides some [**utility functions**](https://pytorch-geometric.readthedocs.io/en/latest/modules/data.html#torch_geometric.data.Data) to infer some basic properties of the underlying graph.


### Creating graphs in PyG (Dummy)


````python
import torch
import numpy as np
import torch_geometric.data as data
# 100 nodes
# feature
features = torch.rand((100, 16), dtype=torch.float) #also called embeddings
print("featyures", features) #dim 16

# Adjacency matrix
rows = np.random.choice(100, 500)
cols = np.random.choice(100, 500)

edges = torch.tensor([rows, cols])


# label
ys = torch.rand((100)).round().long()

edges_attr = np.random.choice(3,500)

graph = data.Data(x=features, edge_index=edges, edge_attr=edges_attr, y=ys)

print(graph)
````

By printing `edge_index`, we can further understand how PyG represents graph connectivity internally.
We can see that for each edge, `edge_index` holds a tuple of two node indices, where the first value describes the node index of the source node and the second value describes the node index of the destination node of an edge.

This representation is known as the **COO format (coordinate format)** commonly used for representing sparse matrices.
Instead of holding the adjacency information in a dense representation $\mathbf{A} \in \{ 0, 1 \}^{|\mathcal{V}| \times |\mathcal{V}|}$, PyG represents graphs sparsely, which refers to only holding the coordinates/values for which entries in $\mathbf{A}$ are non-zero.

# Homophily

Homophily implies that nodes share common property, hobbies etc. 

It is based on **two important notions:**

- Connections between nodes with **same** label
- Connections between nodes with **opposite** labels

To understand homophily, we need to the notion of Network connectance

**Definition 1: Network connectance $p$ :** Measure of the total connections between all nodes in the graph.

$p = \frac{2*num\_edges}{num\_nodes(num\_nodes - 1)}$

num_nodes = 10
num_edges = 14

p = 0.31


**Definition 2- Dyadicity:** Measure of the connectedness between nodes with the **same label** compared to what is expected in a random configuration of the network

$D = \frac{\text{number of same label edges}}{\mathcal{E_{sl}}}$

**Expected number of same label edges $\mathcal{E_{sl}}$** $= \frac{num_l(num_l-1)}{2} \cdot {p}$ \\
where $l$ is the label 

Left graph wrt red node: D = 6/3.1 = 1.9

Righ graph wrt red node: D = 0/3.1 = 0

$D>1$ ==> Dyadic

$D \simeq 1$ ==> Random

$D < 1$ ==> Anti-dyadic

image.png

**Definition 3- Heterophily (H):** Measure of the connectedness between nodes with the **different label** compared to what is expected in a random configuration of the network

$H = \frac{\text{number of cross label edges}}{\mathcal{E_{cl}}}$

**Expected number of cross label edges $\mathcal{E_{cl}}$** $= n_b \cdot n_r \cdot p $ \\
where $l$ is the label 

Left graph: H = 3/7.75 = 0.38

Righ graph: H = 11/7.75 = 1.41

$H > 1$ ==> Heterophilic

$H \simeq 1$ ==> Random

$H < 1$ ==> Heterophobic / Homophily
